```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Replace 'path_to_chrome_driver' with the path to your Chrome driver
driver = webdriver.Chrome(executable_path='path_to_chrome_driver/chromedriver')

# Open the website
driver.get('https://www.openai.com/blog/')

# Find all the blog titles
titles = driver.find_elements(By.XPATH, "//h2[@class='post-preview-title']")

# Extract and print the titles
for title in titles:
    print(title.text)

# Close the browser
driver.quit()
```
